```python


import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import linear_model
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report

# Leer los datos desde el archivo CSV
data = pd.read_csv('C:/Users/TATIANA/anaconda3/framingham.csv', sep= ",")
data.head()
# Eliminar filas con datos faltantes o valores no numéricos
data.dropna(inplace=True)


# Separar los datos de entrada (X) y la variable de salida (y)
X = data[['male', 'age', 'education', 'currentSmoker', 'cigsPerDay', 'BPMeds', 'prevalentStroke', 'prevalentHyp', 'diabetes', 'totChol', 'sysBP', 'diaBP', 'BMI', 'heartRate', 'glucose']]
y = data['TenYearCHD']
# Crear un objeto de modelo de regresión logística con un número máximo de iteraciones mayor
model = LogisticRegression(max_iter=1000)

# Entrenar el modelo con los datos de entrada y la variable de salida
model.fit(X, y)

# Realizar predicciones utilizando el modelo entrenado
y_pred = model.predict(X)

# Crear una matriz de confusión para evaluar la calidad del modelo
cm = confusion_matrix(y, y_pred)
print(cm)

# Crear un informe de clasificación para evaluar la calidad del modelo
print(classification_report(y, y_pred))
```
